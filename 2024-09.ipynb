```python
def add_file(file_id, number_of_files):
    file = []
    for i in range(number_of_files):
        file.append(file_id)
    return file
```

```python
def build_amphipod(input_array):
    amphipod = []
    id = 0

    # Build the amphipod.
    for i in range(len(input_array)):
        
        # Files are even.
        if (i % 2 == 0):
            amphipod.extend(add_file(id, int(input_array[i])))
            id += 1
        
        # Free space are odd. 
        else:
            amphipod.extend(add_file('.', int(input_array[i])))
    
    return amphipod
```

```python
def amphipod_rearrange(amphipod):

    # Rearrange from the last to the first. 
    for i in range(1,len(amphipod)+1,1):
        idx = i * -1

        if (amphipod[idx] != '.'):
            for n in range(len(amphipod)+idx):
                if (amphipod[n] == '.'):
                    amphipod[n] = int(amphipod[idx])
                    amphipod[idx] = '.'
                    break
    
    return amphipod
    
```

```python
def amphipod_checksum(amphipod):
    checksum = 0

    for i in range(len(amphipod)):
        if (amphipod[i] != '.'):
            checksum += i * int(amphipod[i])
    
    return checksum
```

```python
# input_array = list(eg_data)                 # Convert input into array.
input_array = list(data[0])
amphipod = build_amphipod(input_array)      # Convert array into amphipod.
print(amphipod)
amphipod = amphipod_rearrange(amphipod)
print(amphipod)
answer_a = amphipod_checksum(amphipod)
answer_a
```

```python
# input_array = list(eg_data)                 # Convert input into array.
input_array = list(data[0])

print(input_array)
```

```python
def build_amphipod_v2(input_array):
    amphipod = []
    id = 0

    # Build the amphipod.
    for i in range(len(input_array)):
        
        # Files are even.
        if (i % 2 == 0):
            amphipod.extend(add_file(id, int(input_array[i])))
            id += 1
        
        # Free space are odd. 
        else:
            amphipod.extend(add_file('.', int(input_array[i])))
    
    return amphipod, id
```

```python
def get_blank_idx(array, length, min_idx):
    blank_idx = []

    for i in range(len(array)-1, min_idx, -1):

        if (array[i] == '.'):
            blank_idx.append(i)
        else:
            blank_idx = []

        if (len(blank_idx) == length):
            return blank_idx
        
    return []
        
```

```python
import numpy as np

amphipod, max_id = build_amphipod_v2(input_array)
amphipod = np.array(amphipod)
amphipod = np.flip(amphipod)

print(max_id)
print(amphipod)
print(np.size(amphipod))
```

```python
for i in range(max_id-1, -1, -1):
    file_loc = np.where(amphipod == str(i))[0]                              # Get file location.
    blank_loc = get_blank_idx(amphipod, np.size(file_loc), file_loc[-1])    # Find blank spaces.

    # print(i, file_loc, blank_loc)

    if (len(blank_loc) > 0):                # If there is space, swap file locations. 
        for n in range(len(blank_loc)):
            amphipod[file_loc[n]], amphipod[blank_loc[n]] = amphipod[blank_loc[n]], amphipod[file_loc[n]]

amphipod = np.flip(amphipod)    # Reverse the array order. 
print(amphipod)

answer_b = amphipod_checksum(amphipod)
print(answer_b)
```
